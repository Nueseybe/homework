### Bayes Rule

Bayes' theorem is used to calculate how the probability of an event changes when another independent event occurs. In this case, the probability that a person will have cancer can be calculated under the conditions that we have information about the person's sex and age. Bayes' theorem can be expressed as:

Formul :   P(A|B) = P(A)* P(B|A) / P(B)

QUESTION

On a four-lane highway, cars either go fast or not. Faster cars should go in the far left lane. At any given time, 20% of the cars are in the leftmost lane. Overall, 40% of cars on the highway are classified as going fast. Of all cars, 90% of those in the leftmost lane are going fast. If a car is going fast, what is the probability that it is in the Leftmost Lane?

```python
#Solution
P(left) = 0,2
P(right) = 0,8
P(fast) = 0,4
P(slow) = 0.6
P(left|fast) = 0,9
P(left|slow) = 0,10
```

```python
# P(fast|left)= P(left|fast)* P(fast) / P(left)

# P(left) = P(left|fast)* P(fast)+ P(left|slow)* P(slow)

left = 0.9 * 0.4 + 0.1 * 0.6
fast_left = (0.9 * 0.4) / left
fast_left
print("If a car is going fast, the probability of being in the leftmost lane is: ",fast_left)
```
