```python
import numpy as np
import pandas as pd
n = 100
prices = np.random.normal(50, 10, n)
return_prob = np.random.uniform(0, 1, n) 
returns = return_prob < 0.2
df = pd.DataFrame({'Price': prices,'Urun_Degerlendirme_Oran':return_prob ,'Return': returns})
df['Price_Category'] = pd.cut(df['Price'], bins=[0, 30, 50, 70, 100])
```

SORULAR

1- Verilen bir ürünün fiyatı 50-70 aralığındaysa, ürünün geri dönüş olasılığı nedir?

```python
df.query("Price_Category == '(50, 70]'")['Return'].value_counts(normalize=True)[True]
```

2- Ürün değerlendirme oranı 0.5'ten büyük olan bir ürün alındığında, fiyatının 30-50 aralığında olma olasılığı nedir?


```python
#P(Price_Category=(30, 50] | Return > 0.5)
df.query("Urun_Degerlendirme_Oran > 0.5")['Price_Category'].value_counts(normalize='(50, 70]')['(50, 70]']
```

3- Bir ürünün fiyatı 50-70 aralığındaysa, ürün değerlendirme oranı 0.5'ten büyük olduğunda geri dönüş olasılığı nedir?

```python
#P(Return=True | Price_Category=(50, 70] ^ Urun_Degerlendirme_Oran > 0.5)
df.query("Price_Category=='(50, 70]' and Urun_Degerlendirme_Oran>0.5")['Return'].value_counts(normalize=True)
```

4- Bir ürünün fiyatı 70'den fazla olduğunda, ürün değerlendirme oranı 0.5'ten büyük olma olasılığı nedir?

```python
#P(Urun_Degerlendirme_Oran>0.5 | Price_Category=(70, 100])
df.query("Price_Category=='(70, 100]'")['Urun_Degerlendirme_Oran'].value_counts("Urun_Degerlendirme_Oran>0.5")
```
