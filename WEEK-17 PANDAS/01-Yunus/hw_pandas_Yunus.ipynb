# 1- Preprocessing

## Open the file

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from pandas import DataFrame, Series

df = DataFrame()
df = pd.read_csv('hw_data1.csv')
```

## Examine file

```python
df.head
```

## Define missing values

```python
df.isnull()
```

## Decide about missing values (fill or drop)

```python
# df = df.dropna(ignore_index=True)


for column in df.columns:

# Here I have changedd nan volues with the most used values of the column
    most_common = df[column].value_counts().index[0]
    df[column] = df[column].fillna(most_common)
df
```

## Why did you fill missing values or if you have dropped, why did you do so? 

```python
## First i have droppd all the missing rows but then I am left with 43 rows in place of 109 rows. I have lost too much data.
## There for i have filled the missing values with mean or most frequently used words.
```

## Find outliers if there are

```python
Q1 = df['Age'].quantile(0.25)
Q3 = df['Age'].quantile(0.75)
IQR = Q3 - Q1
outliers = df[(df['Age'] < Q1 - 1.5*IQR) | (df['Age'] > Q3 + 1.5*IQR)]
outliers
```

## Decide what will you do with outliers

```python
# I have changed the otliers with the mean of non outliers to save other datas
mean_nonoutliers=df.loc[df['Age'] != 500, 'Age'].mean()
df.loc[df['Age'] == 500, 'Age'] =mean_nonoutliers
df.loc[:,'Age']
```

## Find duplicates, how many are there? Decide what will you do about that?

```python
df.duplicated().sum()
```

```python


df =df.drop_duplicates()
df=df.reset_index(drop=True)
df
```

# 2- Categorization

## Categorize satisfaction in 3 level and label them as 'Low', 'Moderate' and 'High'. Add a new column named 'Satisfaction Category' and store them there

#### ||| Low Satisfaction : (1-3) ||| Moderate Satisfaction : (4-7) ||| High Satisfaction : (8-10) |||

```python
df['Satisfaction_Category']=pd.cut(df['Satisfaction Level'], bins=[1, 4, 8, 10],labels=['Low', 'Modarate', 'High'])
df['Satisfaction_Category']
df
```

## Categorize income in 3 level and label them as 'Low', 'Moderate' and 'High'. Add a new column named 'Income Category' and store them there

```python
df['Income'].max()
df['Income'].min()
df['Income_Category']=pd.cut(df['Income'], bins=[0, 30000, 45000, 100000],labels=['Low', 'Modarate', 'High'])
df['Income_Category']

```
