```python
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

sns.set_style("darkgrid")
```

```python
data_origin = pd.read_csv("120_years_of_olympics_data.csv")
df = data_origin.copy()
df.head()
```

```python
def find_uniques(data):
    for i in data.columns:
        print(f"Uniques values of df[{i}] are :",end=(f"\n{20*'*'}\n"))
        print(df[i].unique())

def find_value_counts(data):
    for i in data.columns:
        print(f"Values counts of df[{i}] are :",end=(f"\n{20*'*'}\n"))
        print(df[i].value_counts())

def find_value_counts_ct(data):
    for i in data.columns:
        print(f"Values counts of df[{i}] are :",end=(f"\n{20*'*'}\n"))
        print(df[i].value_counts().count())
```

```python
find_uniques(df)
```

```python
find_value_counts_ct(df)
```

```python
find_value_counts(df)
```

```python
df.isnull().sum()
```

```python
def fill_with_mean(data,column):
    data[column] = data[column].fillna(data[column].mean())
    print("Succesfully filled missing values with mean- Missing values :",data[column].isnull().sum())
```

```python
fill_with_mean(df,"Age")
```

```python
fill_with_mean(df,"Height")
```

```python
fill_with_mean(df,"Weight")
```
