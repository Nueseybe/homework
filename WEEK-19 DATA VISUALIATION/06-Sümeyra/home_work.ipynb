```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
titanic = sns.load_dataset('titanic')
df=titanic.copy()
df.head()
```

'survived': Survival status (0 = Not survived, 1 = Survived)
'pclass': Ticket class (1 = 1st class, 2 = 2nd class, 3 = 3rd class)
'sex': Gender
'age': Age
'sibsp': Number of siblings/spouses aboard the ship
'parch': Number of parents/children aboard the ship
'fare': Fare (ticket price)
'embarked': Port of embarkation (C = Cherbourg, Q = Queenstown, S = Southampton)
'class': Passenger class
'who': Indicates if the person is a male, female, or child
'adult_male': Indicates if the person is an adult male or not (boolean value)
'deck': Deck where the cabin is located
'embark_town': Full name of the port of embarkation
'alive': Indicates if the person is alive or not
'alone': Indicates if the person is traveling alone or not (boolean value)

'survived': Hayatta kalma durumu (0 = Hayatta kalmadı, 1 = Hayatta kaldı)

'sex': Cinsiyet
'age': Yaş
'sibsp': Gemiye aynı anda binen kardeş/eş sayısı
'parch': Gemiye aynı anda binen ebeveyn/çocuk sayısı
'fare': Bilet ücreti
'embarked': Biniş limanı (C = Cherbourg, Q = Queenstown, S = Southampton)
'class': Yolcu sınıfı
'who': Kişinin erkek, kadın veya çocuk olduğunu gösteren bir kategorik değer
'adult_male': Kişinin yetişkin erkek olup olmadığını gösteren bir boolean değer
'deck': Kabinin bulunduğu güverte
'embark_town': Biniş limanının tam adı
'alive': Hayatta olup olmadığını gösteren bir kategorik değer
'alone': Yalnız seyahat edip etmediğini gösteren bir boolean değer
